
#### Balanceo de clases y análisis de potencia

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import re

path_data = "../processed_data/feature_data/"
target_subjects = ['019', '020'] # Los más significativos

# 1. Agrupar archivos por ID
all_files = [f for f in os.listdir(path_data) if f.endswith('.csv')]
subjects_dict = {}
for f in all_files:
    match = re.search(r'subject_(\d+)', f)
    if match:
        sid = match.group(1)
        if sid in target_subjects:
            if sid not in subjects_dict: subjects_dict[sid] = []
            subjects_dict[sid].append(f)

power_features = ['alpha_power_Pz', 'alpha_power_PO7', 'theta_power_PO8']
ispc_features = ['ispc_PO7_PO8_alpha_signal', 'ispc_Pz_PO8_alpha_signal', 'ispc_PO7_Pz_theta_signal']

analysis_data = []

print("Analizando sujetos significativos...")

for sid, files in subjects_dict.items():
    # Unificar sesiones
    df = pd.concat([pd.read_csv(os.path.join(path_data, f)) for f in files], ignore_index=True)

    # Feature Engineering: Convertir potencias a dB
    for feat in power_features:
        df[f'{feat}_dB'] = 10 * np.log10(df[f'{feat}_signal'] / (df[f'{feat}_baseline'] + 1e-6))

    # Guardar conteo para balanceo
    counts = df['label'].value_counts()
    print(f"\nID {sid}: Clase 0 (On-task) = {counts.get(0,0)} | Clase 1 (MW) = {counts.get(1,0)}")

    # Preparar para plot
    selected_cols = [f'{f}_dB' for f in power_features] + ispc_features
    df_long = df.melt(id_vars=['label'], value_vars=selected_cols,
                      var_name='Feature', value_name='Value')
    df_long['subject_id'] = f"Sujeto {sid}"
    analysis_data.append(df_long)

# 2. Visualización
combined_df = pd.concat(analysis_data)

plt.figure(figsize=(16, 7))
sns.set_theme(style="white")

# Usamos un FacetGrid para ver a los dos sujetos por separado
g = sns.catplot(
    data=combined_df, kind="bar",
    x="Feature", y="Value", hue="label",
    col="subject_id", palette="magma",
    height=5, aspect=1.5, capsize=.1, errorbar=('ci', 95)
)

g.set_axis_labels("", "Valor (dB para Potencia / Ratio para ISPC)")
g.set_xticklabels(rotation=45)
g.set_titles("{col_name}")
plt.subplots_adjust(top=0.85)
g.fig.suptitle('Diferencias de Señal en Sujetos con Mayor Significancia (S1 + S2)', fontsize=16)

plt.show()

# 3. Resumen de medias por sujeto
for sid in target_subjects:
    subj_data = combined_df[combined_df['subject_id'] == f"Sujeto {sid}"]
    summary = subj_data.groupby(['Feature', 'label'])['Value'].mean().unstack()
    summary['Diff'] = summary[1] - summary[0]
    print(f"\n--- Resumen Sujeto {sid} ---")
    print(summary)
```

```python
def check_class_balance(subjects_to_check, path_data):
    balance_results = []

    for sid in subjects_to_check:
        # Buscamos los archivos del sujeto
        files = [f for f in os.listdir(path_data) if f"subject_{sid}" in f]
        df = pd.concat([pd.read_csv(os.path.join(path_data, f)) for f in files])

        counts = df['label'].value_counts().sort_index()
        n_total = len(df)

        balance_results.append({
            'Sujeto': sid,
            'Total Epocas': n_total,
            'On-Task (0)': counts.get(0, 0),
            'Mind-Wandering (1)': counts.get(1, 0),
            '% Clase Minoritaria': (counts.min() / n_total) * 100
        })

    return pd.DataFrame(balance_results)


# Ejecución para tus sujetos significativos
df_balance = check_class_balance(['002', '010', '018', '001', '019','007'], "../processed_data/feature_data/")
print(df_balance)
```
